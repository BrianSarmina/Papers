```python
################################################### COVARIANCE FRACTION ###############################################################
def cov_fraction(A):
    """Offdiag L1 norm divided by sum of diag entries."""
    off = np.sum(np.abs(A)) - np.sum(np.abs(np.diag(A)))
    diag = np.sum(np.abs(np.diag(A)))
    return off/diag if diag != 0 else 0.0

Ns = [4, 7, 10]
depths = [2, 4, 6]   # corresponding to your 3, 6, 9parameter lists

def plot_case(non_list, ent_list, title):
    plt.figure(figsize=(6,4))
    styles = {
        'non-ent': {'linestyle': '-',  'marker': 'o'},
        'ent':     {'linestyle': '--', 'marker': 's'}
    }
    for matrices, label_prefix in [(non_list, "non-ent"), (ent_list, "ent")]:
        style = styles[label_prefix]
        for p_idx, mat_group in enumerate(matrices):
            rs = [cov_fraction(A) for A in mat_group]
            plt.plot(
                Ns, rs,
                linestyle=style['linestyle'],
                marker=style['marker'],
                label=f"{label_prefix}, p={depths[p_idx]}"
            )
    plt.title(title, fontsize=30)
    plt.xlabel("Number of nodes", fontsize=26)
    plt.ylabel(r"Covariance fraction $r$", fontsize=26)
    plt.legend(fontsize=17)
    plt.grid(True)

# two separate plots
plot_case(list_of_matrices_cyclic_non, list_of_matrices_cyclic_ent, "Cyclic maxcut: covariance fraction")
plot_case(list_of_matrices_complete_non, list_of_matrices_complete_ent, "Complete maxcut: covariance fraction")

plt.show()
##########################################################################################################################################
```

## Trace Inverse test calcultation

```python
def get_trace_and_trace_inv(matrices):
    """Return Tr(F) and Tr(F^{-1}) for a list of QFI matrices.

    We symmetrize each matrix before inverting, since numerically estimated
    QFI matrices can deviate slightly from perfect symmetry.
    """
    traces_F = []
    traces_Finv = []

    for M in matrices:
        # Symmetrize to remove tiny numerical asymmetries
        F = 0.5 * (M + M.T)

        # Invert F; if ever singular/ill-conditioned, you can swap to pinv
        try:
            Finv = np.linalg.inv(F)
        except np.linalg.LinAlgError:
            # fallback (MoorePenrose pseudoinverse)
            Finv = np.linalg.pinv(F)

        traces_F.append(np.trace(F))
        traces_Finv.append(np.trace(Finv))

    return traces_F, traces_Finv
```

```python
### PARAMETERS for analyzed problems ###
N_values = [4, 7, 10]                 # number of nodes
depth_labels = ["2p", "4p", "6p"]     # parameter counts for RX-RY
depth_colors = {"2p": "blue", "4p": "green", "6p": "red"}

def get_eigenvalues(matrices):
    """Return max and min eigenvalues for each QFI matrix in a list."""
    highest_eigenvalues = []
    lowest_eigenvalues = []
    for M in matrices:
        # Symmetrize for numerical safety
        F = 0.5 * (M + M.T)
        eigvals = np.linalg.eigvals(F)
        highest_eigenvalues.append(np.max(eigvals).real)
        lowest_eigenvalues.append(np.min(eigvals).real)
    return highest_eigenvalues, lowest_eigenvalues


def get_trace_and_trace_inv(matrices):
    """Return Tr(F) and Tr(F^{-1}) for each QFI matrix in a list."""
    traces_F = []
    traces_Finv = []
    for M in matrices:
        F = 0.5 * (M + M.T)
        try:
            Finv = np.linalg.inv(F)
        except np.linalg.LinAlgError:
            # fallback if ill-conditioned
            Finv = np.linalg.pinv(F)
        traces_F.append(np.trace(F).real)
        traces_Finv.append(np.trace(Finv).real)
    return traces_F, traces_Finv


def plot_eigs_for_family(list_of_matrices, title_prefix, filename_prefix):
    """
    list_of_matrices: [matrices_3p, matrices_6p, matrices_9p],
                      each matrices_* is [F_N=4, F_N=7, F_N=10]
    """
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    # Left: max & min eigenvalues
    for depth_label, matrices in zip(depth_labels, list_of_matrices):
        color = depth_colors[depth_label]
        max_eigs, min_eigs = get_eigenvalues(matrices)

        ax[0].plot(N_values, max_eigs, marker="o",
                   color=color, label=f"ME-{depth_label}")
        ax[0].plot(N_values, min_eigs, marker="o", linestyle="dotted",
                   color=color, label=f"LE-{depth_label}")

    Ns_dense = list(range(min(N_values), max(N_values) + 1))
    ax[0].plot(Ns_dense, [4 * N for N in Ns_dense],
               linestyle="dashed", color="black", label="Linear 4N")
    ax[0].set_xlabel("Number of nodes", fontsize=14)
    ax[0].set_ylabel("Eigenvalue", fontsize=14)
    ax[0].set_title(f"{title_prefix}: max/min eigenvalues", fontsize=16)
    ax[0].grid(True)
    ax[0].legend(fontsize=10)

    # Right: only max eigenvalues (cleaner view)
    for depth_label, matrices in zip(depth_labels, list_of_matrices):
        color = depth_colors[depth_label]
        max_eigs, _ = get_eigenvalues(matrices)
        ax[1].plot(N_values, max_eigs, marker="o",
                   color=color, label=f"ME-{depth_label}")

    ax[1].plot(Ns_dense, [4 * N for N in Ns_dense],
               linestyle="dashed", color="black", label="Linear 4N")
    ax[1].set_xlabel("Number of nodes", fontsize=14)
    ax[1].set_ylabel("Max eigenvalue", fontsize=14)
    ax[1].set_title(f"{title_prefix}: max eigenvalues", fontsize=16)
    ax[1].grid(True)
    ax[1].legend(fontsize=10)

    plt.tight_layout()
    plt.savefig(f"{filename_prefix}_eigs.png", dpi=600)
    plt.savefig(f"{filename_prefix}_eigs.eps")
    plt.close(fig)


def plot_traces_for_family(list_of_matrices, title_prefix, filename_prefix):
    """
    list_of_matrices: [matrices_3p, matrices_6p, matrices_9p],
                      each matrices_* is [F_N=4, F_N=7, F_N=10]
    """
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    for depth_label, matrices in zip(depth_labels, list_of_matrices):
        color = depth_colors[depth_label]
        traces_F, traces_Finv = get_trace_and_trace_inv(matrices)

        ax[0].plot(N_values, traces_F, marker="o",
                   color=color, label=depth_label)
        ax[1].plot(N_values, traces_Finv, marker="o",
                   color=color, label=depth_label)

    ax[0].set_xlabel("Number of nodes", fontsize=14)
    ax[0].set_ylabel(r"$\mathrm{Tr}(F)$", fontsize=14)
    ax[0].set_title(f"{title_prefix}: Tr(F)", fontsize=16)
    ax[0].grid(True)
    ax[0].legend(title="Depth", fontsize=10)

    ax[1].set_xlabel("Number of nodes", fontsize=14)
    ax[1].set_ylabel(r"$\mathrm{Tr}(F^{-1})$", fontsize=14)
    ax[1].set_title(f"{title_prefix}: Tr($F^{{-1}}$)", fontsize=16)
    ax[1].grid(True)
    ax[1].legend(title="Depth", fontsize=10)

    plt.tight_layout()
    plt.savefig(f"{filename_prefix}_traces.png", dpi=600)
    plt.savefig(f"{filename_prefix}_traces.eps")
    plt.close(fig)

####################### RX only ###########################
# Cyclic graph, RX, non-entangled mixers
# plot_eigs_for_family(list_of_matrices_cyclic_non, "Cyclic RX non-entangled", "cyclic_rx_non")
plot_traces_for_family(list_of_matrices_cyclic_non, "Cyclic RX non-entangled", "cyclic_rx_non")

# Cyclic graph, RX, entangled mixers
# plot_eigs_for_family(list_of_matrices_cyclic_ent, "Cyclic RX entangled", "cyclic_rx_ent")
plot_traces_for_family(list_of_matrices_cyclic_ent, "Cyclic RX entangled", "cyclic_rx_ent")

# Complete graph, RX, non-entangled mixers
# plot_eigs_for_family(list_of_matrices_complete_non, "Complete RX non-entangled", "complete_rx_non")
plot_traces_for_family(list_of_matrices_complete_non, "Complete RX non-entangled", "complete_rx_non")

# Complete graph, RX, entangled mixers
# plot_eigs_for_family(list_of_matrices_complete_ent, "Complete RX entangled", "complete_rx_ent")
plot_traces_for_family(list_of_matrices_complete_ent, "Complete RX entangled", "complete_rx_ent")

################### RX-RY ##############################
# # Cyclic graph, RX-RY, non-entangled mixers
# # plot_eigs_for_family(list_of_matrices_cyclic_non, "Cyclic RX-RY non-entangled", "cyclic_rxry_non")
# plot_traces_for_family(list_of_matrices_cyclic_non, "Cyclic RX-RY non-entangled", "cyclic_rxry_non")

# # Cyclic graph, RX-RY, entangled mixers
# # plot_eigs_for_family(list_of_matrices_cyclic_ent, "Cyclic RX-RY entangled", "cyclic_rxry_ent")
# plot_traces_for_family(list_of_matrices_cyclic_ent, "Cyclic RX-RY entangled", "cyclic_rxry_ent")

# # Complete graph, RX-RY, non-entangled mixers
# # plot_eigs_for_family(list_of_matrices_complete_non, "Complete RX-RY non-entangled", "complete_rxry_non")
# plot_traces_for_family(list_of_matrices_complete_non, "Complete RX-RY non-entangled", "complete_rxry_non")

# # Complete graph, RX-RY, entangled mixers
# # plot_eigs_for_family(list_of_matrices_complete_ent, "Complete RX-RY entangled", "complete_rxry_ent")
# plot_traces_for_family(list_of_matrices_complete_ent, "Complete RX-RY entangled", "complete_rxry_ent")
```

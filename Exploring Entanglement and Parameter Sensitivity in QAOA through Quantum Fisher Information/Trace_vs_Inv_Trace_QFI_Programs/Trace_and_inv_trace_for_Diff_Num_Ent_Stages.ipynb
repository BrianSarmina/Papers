```python
def traces_F_and_Finv(mats):
    """Return arrays of Tr(F) and Tr(F^{-1}) for a list of QFI matrices."""
    F_tr = []
    Finv_tr = []
    for M in mats:
        # Tr(F)
        F_tr.append(np.trace(M))
        # Tr(F^{-1}) using eigenvalues (more stable than np.linalg.inv)
        w, _ = np.linalg.eigh(M)
        Finv_tr.append(np.sum(1.0 / w))
    return np.array(F_tr), np.array(Finv_tr)

# Parameters considering the analyzed problem.
N = 7
ent_stages = np.array([1, 2, 3])  # 1L, 2L, 3L entanglement stages
linear_scaling = 4 * N
heisenberg_lim = 4 * (N ** 2)
max_scaling_rx = 4*(N**2 + N)
max_scaling_rxry = 4*(N**2 + 2*N)

### Group matrices pattern RX ###
# qfi_rx = [CYC-ENT 1D, COM-ENT 1D, CYC-ENT 2D, COM-ENT 2D, CYC-ENT 3D, COM-ENT 3D]
mats_rx_cyc = qfi_rx[0::2]   # indices 0,2,4
mats_rx_com = qfi_rx[1::2]   # indices 1,3,5

F_rx_cyc, Finv_rx_cyc = traces_F_and_Finv(mats_rx_cyc)
F_rx_com, Finv_rx_com = traces_F_and_Finv(mats_rx_com)

### Group matrices pattern RX-RY ###
# qfi_rxry = [CYC-ENT 1D, COM-ENT 1D, CYC-ENT 2D, COM-ENT 2D, CYC-ENT 3D, COM-ENT 3D]
mats_rxry_cyc = qfi_rxry[0::2]
mats_rxry_com = qfi_rxry[1::2]

F_rxry_cyc, Finv_rxry_cyc = traces_F_and_Finv(mats_rxry_cyc)
F_rxry_com, Finv_rxry_com = traces_F_and_Finv(mats_rxry_com)

# =====================================================
# RX mixer – sensitivity and estimability
# =====================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)

# --- Tr(F) ---
ax = axes[0]
ax.plot(ent_stages, F_rx_cyc, 'o-', label='RX, CYC-ENT')
ax.plot(ent_stages, F_rx_com, 's--', label='RX, COM-ENT')
# Optional reference lines:
ax.axhline(linear_scaling, linestyle='--', color='k', alpha=0.8, label=r'$4N$')
ax.axhline(heisenberg_lim, linestyle=':', color='k', alpha=0.8, label=r'$4N^2$')
ax.axhline(max_scaling_rx, linestyle='-.',  color='k', alpha=0.8, label=r'$4(N^2 + N)$')
ax.set_xlabel('Number of entanglement stages', fontsize=15)
ax.set_ylabel(r'$\mathcal{F} = \mathrm{Tr}(F)$', fontsize=15)
ax.set_title(r'RX mixer: $\mathcal{F}$ vs entanglement stages', fontsize=17)
ax.grid(True, alpha=0.3)
# ax.legend()
ax.legend(loc="upper left")

# --- Tr(F^{-1}) ---
ax = axes[1]
ax.plot(ent_stages, Finv_rx_cyc, 'o-', label='RX, CYC-ENT')
ax.plot(ent_stages, Finv_rx_com, 's--', label='RX, COM-ENT')
ax.set_xlabel('Number of entanglement stages', fontsize=15)
ax.set_ylabel(r'$\mathcal{F}^{-1} = \mathrm{Tr}(F^{-1})$', fontsize=15)
ax.set_title(r'RX mixer: $\mathcal{F}^{-1}$ vs entanglement stages', fontsize=17)
ax.grid(True, alpha=0.3)
ax.legend()

fig.tight_layout()
plt.savefig("rx_trace_vs_inv_trace_of_diff_num_ent_stages.png", dpi=600)
plt.savefig("rx_trace_vs_inv_trace_of_diff_num_ent_stages.eps")
plt.show()

# =====================================================
#  RX-RY mixer – sensitivity and estimability
# =====================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)

# --- Tr(F) ---
ax = axes[0]
ax.plot(ent_stages, F_rxry_cyc, 'o-', label='RX-RY, CYC-ENT')
ax.plot(ent_stages, F_rxry_com, 's--', label='RX-RY, COM-ENT')
ax.axhline(linear_scaling, linestyle='--', color='k', alpha=0.5, label=r'$4N$')
ax.axhline(heisenberg_lim, linestyle=':', color='k', alpha=0.5, label=r'$4N^2$')
ax.axhline(max_scaling_rxry, linestyle='-.',  color='k', alpha=0.8, label=r'$4(N^2 + 2N)$')
ax.set_xlabel('Number of entanglement stages', fontsize=15)
ax.set_ylabel(r'$\mathcal{F} = \mathrm{Tr}(F)$', fontsize=15)
ax.set_title(r'RX-RY mixer: $\mathcal{F}$ vs entanglement stages', fontsize=17)
ax.grid(True, alpha=0.3)
ax.legend()

# --- Tr(F^{-1}) ---
ax = axes[1]
ax.plot(ent_stages, Finv_rxry_cyc, 'o-', label='RX-RY, CYC-ENT')
ax.plot(ent_stages, Finv_rxry_com, 's--', label='RX-RY, COM-ENT')
ax.set_xlabel('Number of entanglement stages', fontsize=15)
ax.set_ylabel(r'$\mathcal{F}^{-1} = \mathrm{Tr}(F^{-1})$', fontsize=15)
ax.set_title(r'RX-RY mixer: $\mathcal{F}^{-1}$ vs entanglement stages', fontsize=17)
ax.grid(True, alpha=0.3)
ax.legend()

fig.tight_layout()
plt.savefig("rxry_trace_vs_inv_trace_of_diff_num_ent_stages.png", dpi=600)
plt.savefig("rxry_trace_vs_inv_trace_of_diff_num_ent_stages.eps")
plt.show()
```

```python
### ùìï and ùìï^{-1} ###
def F_and_Finv_from_list(matrices):
    """
    For each QFI matrix F in 'matrices' (real, symmetric, PSD),
    return two lists:
        ùìï  = Tr(F)       = sum_i Œª_i
        ùìï^{-1} = Tr(F^{-1}) = sum_i 1/Œª_i
    using the eigenvalues Œª_i.
    """
    F_list = []
    Finv_list = []
    for M in matrices:
        w = np.linalg.eigvalsh(M)
        # Small numerical guard to avoid division by ~0
        w = np.clip(w, 1e-12, None)
        F_list.append(np.sum(w))
        Finv_list.append(np.sum(1.0 / w))
    return np.array(F_list), np.array(Finv_list)


def plot_pattern_comparison(p_list,
                            F_cyc, Finv_cyc,
                            F_comp, Finv_comp,
                            title_prefix,
                            N=7,
                            save_as=None):
    """
    Compare cyclic-entanglement vs complete-entanglement patterns
    for fixed graph/connectivity and mixer type.

    p_list      : list of depths p
    F_cyc,Finv_cyc   : ùìï and ùìï^{-1} for the cyclic-entanglement pattern
    F_comp,Finv_comp : ùìï and ùìï^{-1} for the complete-entanglement pattern
    title_prefix     : text prefix for the panel titles
    """
    linear_scaling = 4 * N
    heisenberg_lim = 4 * (N**2)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # --- ùìï = Tr(F) ---
    ax = axes[0]
    ax.plot(p_list, F_cyc,  "o-", label="Cyclic ent.")
    ax.plot(p_list, F_comp, "o-", label="Complete ent.")
    ax.axhline(linear_scaling, linestyle="--", alpha=0.7, label=r"$4N$")
    ax.axhline(heisenberg_lim, linestyle=":", alpha=0.7, label=r"$4N^{2}$")
    ax.set_xlabel(r"Depth $p$")
    ax.set_ylabel(r"$\mathcal{F} = \mathrm{Tr}(F)$")
    ax.set_title(title_prefix + r": $\mathcal{F}$")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="upper left")

    # --- ùìï^{-1} = Tr(F^{-1}) ---
    ax = axes[1]
    ax.plot(p_list, Finv_cyc,  "o-", label="Cyclic ent.")
    ax.plot(p_list, Finv_comp, "o-", label="Complete ent.")
    ax.set_xlabel(r"Depth $p$")
    ax.set_ylabel(r"$\mathcal{F}^{-1} = \mathrm{Tr}(F^{-1})$")
    ax.set_title(title_prefix + r": $\mathcal{F}^{-1}$")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="upper right")

    fig.tight_layout()
    if save_as is not None:
        plt.savefig(save_as, dpi=300, bbox_inches="tight")
    plt.show()


### ùìï and ùìï^{-1} for each family of QAOA models ###

# RX-only, 7-node cyclic cost graph
F_rx_cyc_cycEnt,  Finv_rx_cyc_cycEnt  = F_and_Finv_from_list(mats_cyc_ent_rx)
F_rx_cyc_fullEnt, Finv_rx_cyc_fullEnt = F_and_Finv_from_list(mats_ent_rx)

# RX-only, 7-node complete cost graph
F_rx_com_cycEnt,  Finv_rx_com_cycEnt  = F_and_Finv_from_list(mats_cyc_ent_rx_com)
F_rx_com_fullEnt, Finv_rx_com_fullEnt = F_and_Finv_from_list(mats_ent_rx_com)

# RX‚ÄìRY, 7-node cyclic cost graph
F_rxry_cyc_cycEnt,  Finv_rxry_cyc_cycEnt  = F_and_Finv_from_list(mats_cyc_ent_rxry_cyc)
F_rxry_cyc_fullEnt, Finv_rxry_cyc_fullEnt = F_and_Finv_from_list(mats_ent_rxry_cyc)

# RX‚ÄìRY, 7-node complete cost graph
F_rxry_com_cycEnt,  Finv_rxry_com_cycEnt  = F_and_Finv_from_list(mats_cyc_ent_rxry_com)
F_rxry_com_fullEnt, Finv_rxry_com_fullEnt = F_and_Finv_from_list(mats_ent_rxry_com)


### Cyclic vs complete entanglement patterns ###

# 1) RX-only, cyclic cost graph (7 nodes)
plot_pattern_comparison(
    p_list=p_cyc_rx,
    F_cyc=F_rx_cyc_cycEnt,
    Finv_cyc=Finv_rx_cyc_cycEnt,
    F_comp=F_rx_cyc_fullEnt,
    Finv_comp=Finv_rx_cyc_fullEnt,
    title_prefix="RX mixer, cyclic graph (7 nodes)"
)

# 2) RX-only, complete cost graph (7 nodes)
plot_pattern_comparison(
    p_list=p_com_rx,
    F_cyc=F_rx_com_cycEnt,
    Finv_cyc=Finv_rx_com_cycEnt,
    F_comp=F_rx_com_fullEnt,
    Finv_comp=Finv_rx_com_fullEnt,
    title_prefix="RX mixer, complete graph (7 nodes)"
)

# 3) RX‚ÄìRY, cyclic cost graph (7 nodes)
plot_pattern_comparison(
    p_list=p_cyc_rxry,
    F_cyc=F_rxry_cyc_cycEnt,
    Finv_cyc=Finv_rxry_cyc_cycEnt,
    F_comp=F_rxry_cyc_fullEnt,
    Finv_comp=Finv_rxry_cyc_fullEnt,
    title_prefix="RX‚ÄìRY mixer, cyclic graph (7 nodes)"
)

# 4) RX‚ÄìRY, complete cost graph (7 nodes)
plot_pattern_comparison(
    p_list=p_com_rxry,
    F_cyc=F_rxry_com_cycEnt,
    Finv_cyc=Finv_rxry_com_cycEnt,
    F_comp=F_rxry_com_fullEnt,
    Finv_comp=Finv_rxry_com_fullEnt,
    title_prefix="RX‚ÄìRY mixer, complete graph (7 nodes)"
)
```

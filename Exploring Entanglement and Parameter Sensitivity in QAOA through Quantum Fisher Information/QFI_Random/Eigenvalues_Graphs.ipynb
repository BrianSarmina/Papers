```python
list_of_matrices_cyclic_non = [matrices_3p_non_cyc, matrices_6p_non_cyc, matrices_9p_non_cyc]
list_of_matrices_cyclic_ent = [matrices_3p_ent_cyc, matrices_6p_ent_cyc, matrices_9p_ent_cyc]
list_of_matrices_complete_non = [matrices_3p_non_com, matrices_6p_non_com, matrices_9p_non_com]
list_of_matrices_complete_ent = [matrices_3p_ent_com, matrices_6p_ent_com, matrices_9p_ent_com]

def get_eigenvalues(matrices):
  highest_eigenvalues = []
  lowest_eigenvalues = []
  for matrix in matrices:
    eigenvalues = np.linalg.eigvals(matrix)
    highest_eigenvalues.append(np.max(eigenvalues))
    lowest_eigenvalues.append(np.min(eigenvalues))
    # print(matrix)
    # print(eigenvalues)
    # print("\n")
  return highest_eigenvalues, lowest_eigenvalues
```

```python
print("Cyclic problems with non-entangled mixing operators")
plot_colors = ['blue', 'green', 'red']
plot_names = ["3p non-ent", "6p non-ent", "9p non-ent"]
count_color = 0
for matrix in list_of_matrices_cyclic_non:
  highest_eigenvalues, lowest_eigenvalues = get_eigenvalues(matrix)
  x_axis = [4, 7, 10] # number of nodes
  # print("Highest: ", highest_eigenvalues)
  # print("Lowest: ", lowest_eigenvalues)
  plt.plot(x_axis, highest_eigenvalues, label=plot_names[count_color], color=plot_colors[count_color])
  plt.plot(x_axis, lowest_eigenvalues, label='Lowest Eigenvalue', color=plot_colors[count_color], linestyle='dotted')
  count_color += 1

plt.plot([x for x in range(4, 10)], [4*(x) for x in range(4, 10)], label="Linear scaling", color='black', linestyle='dashed')
plt.xlabel("Number of nodes")
plt.ylabel("Eigenvalue")
plt.grid()
plt.legend()
  #plt.plot([x for x in range(4, 10)], [4*(x) for x in range(4, 10)], label="Heisenberg limit", color='black', linestyle='dashed')
```
